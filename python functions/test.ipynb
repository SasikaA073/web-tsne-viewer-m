```python
import numpy as np

# Assuming the following constants are defined:
n_cols = 20
n_rows = 15

# TSNE Values between 0 and 1 
tsne_np_arr = np.random.rand(n_rows * n_cols, 2)

# Functions to test
def get_img_id(row: int, col: int):
    return (row - 1) * n_cols + col

def get_img_row_col(img_id: int):
    row = (img_id // n_cols) if img_id % n_cols == 0 else (img_id // n_cols) + 1
    col = img_id % n_cols if img_id % n_cols != 0 else n_cols
    return row, col

def get_img_tsne(tsne_np_arr: np.array, img_id: int):
    return tsne_np_arr[img_id-1]
```

```python
# first image 1,1 
assert get_img_id(row=1, col=1) == 1
assert get_img_row_col(1) == (1, 1)

assert get_img_tsne(tsne_np_arr=tsne_np_arr, img_id=   1).all() == tsne_np_arr[0].all()
```

```python
# last image 15row, 20 col
assert get_img_id(row=15, col=20) == 300
assert get_img_row_col(img_id=300) == (15, 20)

assert get_img_tsne(tsne_np_arr=tsne_np_arr, img_id=300).all() == tsne_np_arr[300-1].all()
```

# Make a new image of 15*20 image



```python
# Load all the image file paths into a list
import os
image_file_paths = os.listdir("test_images")
```

```python
len(image_file_paths)
```

```python
import matplotlib.pyplot as plt
image_file_shapes = []
for image_file_path in image_file_paths:
    image_file_shapes.append(plt.imread("test_images/" + image_file_path).shape)
```

```python
image_file_shapes
```

```python
# Center crop the images to 128x128 pixels and save to folder "test_images_center_cropped" with the 1-index image id as the filename
import cv2
import os

os.makedirs("test_images_center_cropped", exist_ok=True)

for i, image_file_path in enumerate(image_file_paths):
    image = plt.imread("test_images/" + image_file_path)
    center_cropped_image = cv2.resize(image, (128, 128))
    plt.imsave("test_images_center_cropped/" + str(i+1) + ".png", center_cropped_image)
```

```python
# Create a zip file for the folder "test_images_center_cropped"
import zipfile

with zipfile.ZipFile("test_images_center_cropped.zip", "w") as zipf:
    for image_file_path in os.listdir("test_images_center_cropped"):
        zipf.write("test_images_center_cropped/" + image_file_path)
```

```python
center_cropped_file_paths = os.listdir("test_images_center_cropped") 
center_cropped_file_paths
```

```python
# sort center cropped file paths in "1.png", "2.png", "3.png", ... order
center_cropped_file_paths.sort(key=lambda x: int(x.split(".")[0]))
center_cropped_file_paths
```

```python
# save a text file of all the center_cropped image file paths
with open("test_img_paths.txt", "w") as f:
    for image_file_path in center_cropped_file_paths:
        f.write("test_images_center_cropped/" + image_file_path + "\n")
```
